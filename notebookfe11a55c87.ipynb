```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/kaggle/input/datasets/aastha24mehta/dashreels/dataset_youtube-scraper_2026-06-08_12-28-19-975.csv')

df['viewCount'] = pd.to_numeric(df['viewCount'], errors='coerce')
df['likes'] = pd.to_numeric(df['likes'], errors='coerce')
df['commentsCount'] = pd.to_numeric(df['commentsCount'], errors='coerce')

df['engagement_rate'] = ((df['likes'] + df['commentsCount']) / df['viewCount'] * 100).round(2)

print("Total videos:", len(df))
print("Average views:", df['viewCount'].mean().round(0))
print("Average engagement rate:", df['engagement_rate'].mean().round(2), "%")
df.head()
```

```python
top10 = df.nlargest(10, 'viewCount')[['title', 'viewCount', 'likes', 'engagement_rate']]

plt.figure(figsize=(12, 6))
sns.barplot(data=top10, x='viewCount', y='title', hue='title', legend=False, palette='magma')
plt.title('Dashreels - Top 10 Most Viewed Videos')
plt.xlabel('Views')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
type_analysis = df.groupby('type').agg(
    avg_views=('viewCount', 'mean'),
    avg_likes=('likes', 'mean'),
    avg_engagement=('engagement_rate', 'mean'),
    total_videos=('title', 'count')
).round(2)

print(type_analysis)

plt.figure(figsize=(8, 5))
sns.barplot(data=df, x='type', y='viewCount', palette='coolwarm')
plt.title('Average Views by Content Type')
plt.xlabel('Content Type')
plt.ylabel('Average Views')
plt.tight_layout()
plt.show()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.to_period('M')

monthly = df.groupby('month').size().reset_index(name='videos_posted')

plt.figure(figsize=(12, 5))
sns.lineplot(data=monthly, x=monthly['month'].astype(str), y='videos_posted', marker='o', color='coral')
plt.title('Dashreels - Videos Posted Per Month')
plt.xlabel('Month')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='duration', y='viewCount', color='purple', alpha=0.6)
plt.title('Video Duration vs Views')
plt.xlabel('Duration (seconds)')
plt.ylabel('Views')
plt.tight_layout()
plt.show()
```
